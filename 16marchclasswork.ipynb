```python
# to find out the greatest number
num=int(input("enter the number:"))
max=0
while(num>0):
  digit=num % 10
  if(digit>max):
    max=digit
  num=num//10
print("the greatest number is:",max)
```

```python
# to display even factors
num=int(input("enter the number:"))
print("the even factors are:")
for i in range(1,num+1):
  if num % i==0:
    if i%2==0:
      print(i)
```
